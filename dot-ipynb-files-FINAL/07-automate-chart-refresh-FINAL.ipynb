```python
### üìà Live Crypto Price Tracker

import requests, pandas as pd, matplotlib.pyplot as plt, time
from datetime import datetime

plt.rcParams["figure.figsize"] = (8,4)

symbol = "BTCUSDT"         # Bitcoin vs USD
interval = 30              # seconds between updates
history = []               # list to hold (time, price)

while True:
    # 1Ô∏è‚É£ fetch current price (Binance public API)
    url = "https://api.binance.us/api/v3/ticker/price?symbol=" + symbol
    res = requests.get(url, timeout=10)
    data = res.json()
    price = float(data["price"])
    now = datetime.now()

    # 2Ô∏è‚É£ store new record
    history.append({"time": now, "price": price})

    # 3Ô∏è‚É£ turn into DataFrame for plotting
    df = pd.DataFrame(history)

    # 4Ô∏è‚É£ clear and redraw
    plt.clf()
    plt.plot(df["time"], df["price"], marker="o")
    plt.title(f"{symbol} price ‚Äî updated {now.strftime('%H:%M:%S')}")
    plt.xlabel("Time")
    plt.ylabel("Price (USD)")
    plt.tight_layout()
    plt.pause(0.01)

    print(f"{now:%H:%M:%S}  ${price:,.2f}")
    time.sleep(interval)
```
