### **df.loc[ ] vs. df.iloc[ ]**
- these are for sub-selecting from df by row and col
- ex: get a 3x3 tic-tac-toe size selection from 8x8 chessboard
- what's the diff?
    - INDIVIDUAL VALUE SELECTION
        - loc[row_name, col_name]
        - iloc[row_num, col_num]
    - RANGE SELECTION
        - loc[start_row_name:end_row_name, start_col_name:end_col_name]
        - iloc[start_row_num:end_row_num, start_col_num:end_col_num]
        - loc end values are inclusive (included in the result)
        - iloc end values are exclusive (not included in the result)
        - 4: means start at 4 and go to the end
        - :4 means start at the beginning and go to 4
        - : means all (so, either all rows or all cols)
        - ::2 means start at beginning and go to end by get every other item
        - ::-1 means start at end and go to the beginning (so, backwards)
            - nums = [1,2,3]
            - rev_nums = nums[::-1] reverse the list

### **Data shaping, apply(lambda)**

We'll be using:

- **lambdas** and apply to do more in-depth data shaping
- notebooks to practice and to explore our data

```python
### Lambdas and Apply

# First, just review "regular" function definitions
# define a function that squares even input number; cubes an odd input
# assume valid input (integer)
def squareOrCubeNum(num):
    # check if number is even: %
    if num % 2 == 0:
        return num ** 2
    else:
        return num ** 3
```

```python
# test the function:
_12_squared = squareOrCubeNum(12)
print(_12_squared)
_11_cubed = squareOrCubeNum(11)
print(_11_cubed)
```

- **ternary expression**: if-else as one-liner
- **syntax:**
    - **return value_X if condition_true else value_Y**

```python
# redo the fuction with the if-else refactored as a one-line ternary
def sqOrCubeNum(n):
        return n**2 if n%2==0 else n**3
```

```python
# test the function:
_12_squared = sqOrCubeNum(12)
print(_12_squared)
_11_cubed = sqOrCubeNum(11)
print(_11_cubed)
```

```python
# lambda version of the above function
# instead of defining a function, declare a variable
# and set that equal to the function which returns a value
# varName = lambda input : output
sqOrCubeN = lambda n : n**2 if n%2==0 else n**3
```

```python
# test the lambda function:
_12_squared = sqOrCubeN(12)
print(_12_squared)
_11_cubed = sqOrCubeN(11)
print(_11_cubed)
```

```python
# What was the most expensive item in an order?
# first, make a new col where price is an actual number
# then, get the max value from that new col
```

```python
# get the max of the 'item_price' col -- even though the prices are all strings
max_item_price = chip_df['item_price'].max()
max_item_price # '$9.39 '
# strings have low-to-order (min-to-max range):
# LOW 0123456789ABCDEFGH...TUVWXYZabcedefgh...tuvwxyz HIGH
# in the string hi-lo range, '9' is greatest digit, so '$9' is greater than '$10' or '$18' or '$850' or '$10000'
```

```python
# test greater than less than with letters:
print(2 > 1)
print('B' > 'A')
print('Z' < 'a')
print('$9.39' > '$1000')
```

```python
# make a new col 'price_num' where the price is number
# to do this, get rid of the '$' and convert str to float
# all this is done iteratively with col.apply(lambda)
# str.replace(find,replace) and returns the new version:
# replace exclamation points with periods:
# str = str.replace('!', '.')
# replace dollar signs with empty strings:
# str = str.replace('$', '')
chip_df['price_num'] = chip_df['item_price'].apply(lambda p : float(p.replace('$','')))
```

```python
print(chip_df.shape) # (4622, 6)
chip_df.head()
```

```python
# most expensive:
highest_price = chip_df['price_num'].max()
print('highest_price:', highest_price) # 44.25
```
