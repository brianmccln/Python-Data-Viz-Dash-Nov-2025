### **df.loc[ ] vs. df.iloc[ ]**
- these are for sub-selecting from df by row and col
- ex: get a 3x3 tic-tac-toe size selection from 8x8 chessboard
- what's the diff?
    - INDIVIDUAL VALUE SELECTION
        - loc[row_name, col_name]
        - iloc[row_num, col_num]
    - RANGE SELECTION
        - loc[start_row_name:end_row_name, start_col_name:end_col_name]
        - iloc[start_row_num:end_row_num, start_col_num:end_col_num]
        - loc end values are inclusive (included in the result)
        - iloc end values are exclusive (not included in the result)
        - 4: means start at 4 and go to the end
        - :4 means start at the beginning and go to 4
        - : means all (so, either all rows or all cols)
        - ::2 means start at beginning and go to end by get every other item
        - ::-1 means start at end and go to the beginning (so, backwards)
            - nums = [1,2,3]
            - rev_nums = nums[::-1] reverse the list
